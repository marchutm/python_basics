Zaimportuj zawarte w plikach dane do notatnika i utwórz na ich podstawie tablicę `recession_data_1` wykonaną według następujących wytycznych:

- Tablica powinna zawierać wyłącznie następujące elementy:</br>
    Indeks: Country Code</br>
        Wiersze powinny być indeksowane trzyliterowymi kodami krajów.</br>
    Kolumny: Country Name, Region Name, Income Group, 2006, 2007, 2008, 2009, 2010, 2011 (w tej kolejności, lata w nagłówkach jako zmienne typu str)</br>
        Kolumna Income Group zawiera informację o stopniu zamożności państwa wg danych z 2008 roku.
        Kolumny 2006-2011 zawierają wartości wskaźnika GDP growth (annual %) dla poszczególnych lat.
- Zmień występujące w kolumnie Income Category kody (H, UM, LM, L) na pełne wyrażenia (High, Upper middle, Lower middle, Low).
- Tablica nie powinna zawierać wierszy, w których znajdują się wartości NaN (w dowolnej kolumnie) oraz wiersza dotyczącego Libii.

Zapisz tablicę do pliku recession_data_1.pkl.

Wyświetl przygotowaną tablicę w taki sposób aby widoczne było po 5 pierwszych i ostatnich wierszy, a zawarte w tablicy liczby (za wyjątkiem lat w nagłówkach kolumn) wyświetlały się z dokładnością do 2 miejsca po przecinku.

```python
# YOUR CODE HERE
data1 = pd.read_csv('raw_data_1.csv', usecols=['Country Code', 'Country Name', '2006 [YR2006]', '2007 [YR2007]',
                                               '2008 [YR2008]', '2009 [YR2009]', '2010 [YR2010]', '2011 [YR2011]'], na_values=["..", "Libya"])
data1.columns = ['Country Name', 'Country Code', '2006', '2007', '2008', '2009', '2010', '2011']
data1.dropna(inplace=True)
data2 = pd.read_csv('raw_data_2.csv', sep=';', usecols=['Region Name', 'ISO-alpha3 Code'], na_values='..')
data2.columns = ['Region Name', 'Country Code']
data2.dropna(inplace=True)
data3 = pd.read_excel('raw_data_3.xlsx', sheet_name='Country Analytical History', skiprows=10, na_values='..')
data3 = data3.iloc[:,[0,23]]
data3.columns = ['Country Code', 'Income Group']
data3.dropna(inplace=True)
data3['Income Group'].replace({'L': 'Low', 'LM': 'Lower middle', 'UM': 'Upper middle', 'H': 'High'}, inplace=True)
recession_data_1 = data1.merge(data2, how='inner', on='Country Code')
recession_data_1 = recession_data_1.merge(data3, how='inner', on='Country Code')
recession_data_1.set_index('Country Code', inplace=True)
recession_data_1 = recession_data_1[['Country Name', 'Region Name', 'Income Group', '2006', '2007', '2008', '2009', '2010', '2011']]
recession_data_1.to_pickle('recession_data_1.pkl')
indexes = recession_data_1.index
recession_data_1_display = recession_data_1.style.format({'2006': '{:.2f}', '2007': '{:.2f}', '2008': '{:.2f}', '2009': '{:.2f}', '2010': '{:.2f}', '2011': '{:.2f}'}).hide(indexes[5:-5])
recession_data_1_display
```
