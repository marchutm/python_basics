#### a) Obliczanie wartości wybranych opisowych charakterystyk rozkładu

Korzystając z danych zawartych w tablicy `recession_table_1` oblicz dla każdego roku następujące opisowe charakterystyki rozkładu:
- wartość średnia (ang. mean),
- odchylenie standardowe (ang. standart deviation),
- kwartyle (ang. quartiles),
- moda (ang. mode).

Zestaw wyniki w tablicy `descriptive_parameters_2`.

> Indeks: </br>
> Zawiera wiersze o następujących nazwach: mean, std, Q1, Q2, Q3, mode.

> Kolumny: lata (osobna kolumna dla każdego roku opatrzona nagłówkiem 2006, 2007 itd. sformatowanym jako str) </br>

Uwaga! W celu otrzymania maksymalnej liczby punktów otrzymane wyniki powinny być w całości wynikiem pracy napisanego kodu, bez manualnego określania, w którym przedziale znajdują się określone statystyki itp.

```python
# YOUR CODE HERE

# mean
mid_points = np.array([(interval.right + interval.left) / 2 for interval in recession_table_1.index])
n_freq = recession_table_1.sum()
ssum = np.dot(mid_points, recession_table_1.values)
mean = ssum / n_freq

# std
# std = mid_points - mean
diff = np.array([mid_points - mn for mn in mean]).T
std = np.sqrt(np.sum((diff**2 * recession_table_1.values), axis=0) / n_freq)

# quartiles
Q1 = []
Q2 = []
Q3 = []
n = recession_table_1['2006'].sum()

for y in recession_table_1.columns:
    q = recession_table_1[y].cumsum()
    index_q1 = q[q > 0.25 * n].index[0]
    index_q2 = q[q > 0.5 * n].index[0]
    index_q3 = q[q > 0.75 * n].index[0]

    q1 = (n/4 - recession_table_1[y].loc[:index_q1-5].sum())*5 / recession_table_1.loc[index_q1, y] + index_q1.left
    Q1.append(q1)

    q2 = (n/2 - recession_table_1[y].loc[:index_q2-5].sum())*5 / recession_table_1.loc[index_q2, y] + index_q2.left
    Q2.append(q2)

    q3 = (3*n/4 - recession_table_1[y].loc[:index_q3-5].sum())*5 / recession_table_1.loc[index_q3, y] + index_q3.left
    Q3.append(q3)

years = recession_table_1.columns
Q1 = pd.Series(Q1, index=years)
Q2 = pd.Series(Q2, index=years)
Q3 = pd.Series(Q3, index=years)
    
# mode
mode = []
for column in recession_table_1.columns:
    max_value_index = recession_table_1[column].idxmax()
    l = max_value_index.left
    h = max_value_index.length
    fm = recession_table_1.at[max_value_index, column]
    previous_index = max_value_index - 5
    following_index = max_value_index + 5
    f1 = recession_table_1.at[previous_index, column]
    f2 = recession_table_1.at[following_index, column]
    mod = l + ((fm - f1)/((fm-f1)+(fm-f2))) * h
    mode.append(mod)

mode = pd.Series(mode, index=years)

descriptive_parameters_2 = pd.DataFrame([mean, std, Q1, Q2, Q3, mode], index=['mean', 'std', 'Q1', 'Q2', 'Q3', 'mode'], columns=['2006', '2007', '2008', '2009', '2010', '2011'])
descriptive_parameters_2
```

```python
### Komórka testowa
# Testy wybranych wartości z tablicy
assert np.isclose(descriptive_parameters_2.loc['mean', '2006'], 5.606061)
assert np.isclose(descriptive_parameters_2.loc['std', '2007'], 4.184001)
assert np.isclose(descriptive_parameters_2.loc['Q1', '2008'], 0.916667)
assert np.isclose(descriptive_parameters_2.loc['Q2', '2009'], 0.220588)
assert np.isclose(descriptive_parameters_2.loc['Q3', '2010'], 7.604167)
assert np.isclose(descriptive_parameters_2.loc['mode', '2011'], 3.691589)
```
