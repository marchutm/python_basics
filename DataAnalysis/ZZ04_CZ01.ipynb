#### a) Wyliczanie prawdopodobieństw na podstawie wartości

Oblicz i zapisz do zmiennych o podanych nazwach następujące wartości prawdopodobieństwa:
- `p1` - $P(X < -1.5)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład t-Studenta o 10 stopniach swobody - $t(10)$,
- `p2` - $P(X > 2)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład F Snedecora o 5 i 50 stopniach swobody - $F(5, 50)$,
- `p3` - $P(3 < X < 10)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład chi-kwadrat o 4 stopniach swobody - $\chi^2(4)$,
- `p4` - $P(X < -2.75 | X > 1.5)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład normalny o wartości oczewiwanej równej -0.5 i odchyleniu standardowym równym 1.5 - $N(-0.5, 1.5)$,

Każde z rozwiązań zilustruj wykresem zawierającym krzywą gęstości prawdopodobieństwa rozkładu z zaznaczonym kolorem fragmentem pola pod krzywą odpowiadający wyznaczonej wartości prawdopodobieństwa. Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:
- ustaw zakres osi x od wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.0001 do wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.9999,
- w tytule wykresu zawrzyj uzupełniony wzór na prawdopodobieństwo uzupełniony o wyliczoną wartość prawdopodobieństwa podaną do 4 miejsca po przecinku (np. P(X < -1.5) = x.xxxx, gdzie x.xxxx to obliczona wartość).

```python
# p1
p1 = sp.stats.t.cdf(-1.5, df=10)

x1_low = sp.stats.t.ppf(0.0001, df=10)
x1_high = sp.stats.t.ppf(0.9999, df=10)

x1 = np.linspace(x1_low, x1_high, 1000)
p1_pdf = sp.stats.t.pdf(x1, 10)

plt.figure(figsize=(14, 10), facecolor='white')
sns.lineplot(x=x1, y=p1_pdf, color='black')
plt.fill_between(x1, p1_pdf, where=(x1 <= -1.5), color='green', alpha=0.5)
plt.title(f'P(X < - 1,5) = {p1:.4f}')
plt.xlabel('x')
plt.ylabel('pdf')
plt.xlim([x1_low, x1_high])
plt.show()
```

```python
# p2
p2 = sp.stats.f.sf(2, dfn=5, dfd=50)

x2_low = sp.stats.f.ppf(0.0001, dfn=5, dfd=50)
x2_high = sp.stats.f.ppf(0.9999, dfn=5, dfd=50)

x2 = np.linspace(x2_low, x2_high, 1000)
p2_pdf = sp.stats.f.pdf(x2, dfn=5, dfd=50)

plt.figure(figsize=(14, 10), facecolor='white')
sns.lineplot(x=x2, y =p2_pdf, color='black')
plt.fill_between(x2, p2_pdf, where=(x2 >= 2), color='green', alpha=0.5)
plt.title(f'P(X > 2) = {p2:.4f}')
plt.xlabel('x')
plt.ylabel('pdf')
plt.xlim([x2_low, x2_high])
plt.show()
```

```python
# p3
p3 = sp.stats.chi2.cdf(10, df=4) - sp.stats.chi2.cdf(3, df=4)

x3_low = sp.stats.chi2.ppf(0.0001, df=4)
x3_high = sp.stats.chi2.ppf(0.9999, df=4)

x3 = np.linspace(x3_low, x3_high, 1000)
p3_pdf = sp.stats.chi2.pdf(x3, df=4)

plt.figure(figsize=(14, 10), facecolor='white')
sns.lineplot(x=x3, y =p3_pdf, color='black')
plt.fill_between(x3, p3_pdf, where=((x3 > 3) & (x3 < 10)), color='green', alpha=0.5)
plt.title(f'P(3 < X < 10) = {p3:.4f}')
plt.xlabel('x')
plt.ylabel('pdf')
plt.xlim([x3_low, x3_high])
plt.show()
```

```python
# p4
p4 = (sp.stats.norm.cdf(-2.75, loc=-0.5, scale=1.5) + (sp.stats.norm.sf(1.5, loc=-0.5, scale=1.5)))

x4_low = sp.stats.norm.ppf(0.0001, loc=-0.5, scale=1.5)
x4_high = sp.stats.norm.ppf(0.9999, loc=-0.5, scale=1.5)

x4 = np.linspace(x4_low, x4_high, 1000)
p4_pdf = sp.stats.norm.pdf(x4, loc=-0.5, scale=1.5)

plt.figure(figsize=(14, 10), facecolor='white')
sns.lineplot(x=x4, y=p4_pdf, color='black')
plt.fill_between(x4, p4_pdf, where=((x4 < -2.75) | (x4 > 1.5)), color='green', alpha=0.5)
plt.title(f'P(X < -2.75 | X>1.5) = {p4:.4f}')
plt.xlabel('x')
plt.ylabel('pdf')
plt.xlim([x4_low, x4_high])
plt.show()
```

#### b) Wyliczanie wartości na podstawie prawdopodobieństw

Rozwiąż i zapisz do zmiennych o podanych nazwach rozwiązania następujących równań:
- `s1` - $P(X < s) = 0.25$, gdzie X to zmienna losowa ciągła posiadająca rozkład chi-kwadrat o 5 stopniach swobody - $\chi^2(5)$,
- `s2` - $P(X > s) = 0.15$, gdzie X to zmienna losowa ciągła posiadająca rozkład F Snedecora o 5 i 20 stopniach swobody - $F(5, 20)$,
- `s3` - $P(-1.25 < X < s) = 0.35$, gdzie x to zmienna losowa ciągła posiadająca rozkład normalny o wartości oczewiwanej równej -0.4 i odchyleniu standardowym równym 0.75 - $N(-0.4, 0.75)$,

Każde z rozwiązań zilustruj wykresem zawierającym krzywą gęstości prawdopodobieństwa rozkładu z zaznaczonym kolorem fragmentem pola pod krzywą odpowiadający wartości prawdopodobieństwa określonej przez rozwiązywane równanie. Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:
- ustaw zakres osi x od wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.0001 do wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.9999,
- oznacz krawędź obszaru, na której znajduje się wyznaczona wartość $s$ linią w wyróżniającym się kolorze,
- w tytule wykresu zawrzyj uzupełniony wzór na prawdopodobieństwo uzupełniony o wyliczoną wartość zmiennej $X$ podaną do 4 miejsca po przecinku (np. P(X < s.ssss) = 0.25, gdzie s.ssss to obliczona wartość).

```python
# s1
s1 = sp.stats.chi2.ppf(0.25, df=5)

xs1_low = sp.stats.chi2.ppf(0.0001, df=5)
xs1_high = sp.stats.chi2.ppf(0.9999, df=5)

xs1 = np.linspace(xs1_low, xs1_high, 1000)
ps1_pdf = sp.stats.chi2.pdf(xs1, df=5)

plt.figure(figsize=(14, 10), facecolor='white')
sns.lineplot(x=xs1, y=ps1_pdf, color='black')
plt.fill_between(xs1, ps1_pdf, where=(xs1 <= s1), color='green', alpha=0.5)
plt.axvline(x=s1, color='red', linestyle='-', ymin=0.045, ymax=0.945)
plt.title(f'P(X < {s1:.4f}) = 0.25')
plt.xlabel('x')
plt.ylabel('pdf')
plt.xlim([xs1_low, xs1_high])
plt.show()
```

```python
# s2
s2 = sp.stats.f.isf(0.15, dfn=5, dfd=20)

xs2_low = sp.stats.f.ppf(0.0001, dfn=5, dfd=20)
xs2_high = sp.stats.f.ppf(0.9999, dfn=5, dfd=20)

xs2 = np.linspace(xs2_low, xs2_high, 1000)
ps1_pdf = sp.stats.f.pdf(s2, dfn=5, dfd=20)

plt.figure(figsize=(14, 10), facecolor='white')
sns.lineplot(x=xs2, y=sp.stats.f.pdf(xs2, dfn=5, dfd=20), color='black')
plt.fill_between(xs2, sp.stats.f.pdf(xs2, dfn=5, dfd=20), where=(xs2 >= s2), color='green', alpha=0.5)
plt.axvline(x=s2, color='red', linestyle='-', ymin=0.045, ymax=0.29)
plt.title(f'P(X > {s2:.4f}) = 0.15')
plt.xlabel('x')
plt.ylabel('pdf')
plt.xlim([xs2_low, xs2_high])
plt.show()
```

```python
# s3
mn = -0.4
std = 0.75

s3 = sp.stats.norm.ppf(0.35 + sp.stats.norm.cdf(-1.25, mn, std), mn, std)

xs3_low = sp.stats.norm.ppf(0.0001, mn, std)
xs3_high = sp.stats.norm.ppf(0.9999, mn, std)

xs3 = np.linspace(xs3_low, xs3_high, 1000)

ps3_pdf = sp.stats.norm.pdf(s3, mn, std)

plt.figure(figsize=(14, 10), facecolor='white')
sns.lineplot(x=xs3, y=sp.stats.norm.pdf(xs3, mn, std), color='black')  # Krzywa gęstości prawdopodobieństwa
plt.fill_between(xs3, sp.stats.norm.pdf(xs3, mn, std), where=((xs3 > -1.25) & (xs3 < s3)), color='green', alpha=0.5)  # Wypełnienie obszaru pod krzywą
plt.axvline(x=s3, color='red', linestyle='-', ymin=0.045, ymax=0.95)  # Oznaczenie linii dla s3
plt.axvline(x=-1.25, color='red', linestyle='-', ymin=0.045, ymax=0.52)  # Oznaczenie linii dla a
plt.title(f'P({-1.25} < X < {s3:.4f}) = {0.35}')
plt.xlabel('x')
plt.ylabel('pdf')
plt.xlim([xs3_low, xs3_high])
plt.show()
```
